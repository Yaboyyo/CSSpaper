```python
import pandas as pd
import re

# Function to clean text by removing tabs, newlines, pipes, and non-printable characters
def clean_text(text):
    if isinstance(text, str):
        # Remove tabs, newlines, pipes, and non-printable characters
        text = re.sub(r'[\t\n\r|]+', ' ', text)
        text = re.sub(r'[^\x20-\x7E]', '', text)  # Remove non-printable characters
        return text
    return text

# Apply the function to all string columns in each dataframe
def clean_dataframe(df):
    for col in df.select_dtypes(include=['object']).columns:
        df[col] = df[col].apply(clean_text)
    return df

# Standardize column names for each dataframe
df_guardian_cleaned = df_guardian_cleaned.rename(columns={'Publication Date': 'Date', 'Headline': 'Headline', 'source': 'Source'})
df_ABC_headlines = df_ABC_headlines.rename(columns={'publish_date': 'Date', 'headline_text': 'Headline', 'source': 'Source'})
df_Pakistan_headlines = df_Pakistan_headlines.rename(columns={'Date': 'Date', 'Headline': 'Headline', 'source': 'Source'})
df_Irish_Times_articles = df_Irish_Times_articles.rename(columns={'publish_date': 'Date', 'headline_text': 'Headline', 'source': 'Source'})
df_India_Times_articles = df_India_Times_articles.rename(columns={'publish_date': 'Date', 'headline_text': 'Headline', 'source': 'Source'})
df_NYT_cleaned = df_NYT_cleaned.rename(columns={'Publication Date': 'Date', 'Headline': 'Headline', 'source': 'Source'})

# Clean all dataframes
df_guardian_cleaned = clean_dataframe(df_guardian_cleaned)
df_ABC_headlines = clean_dataframe(df_ABC_headlines)
df_Pakistan_headlines = clean_dataframe(df_Pakistan_headlines)
df_Irish_Times_articles = clean_dataframe(df_Irish_Times_articles)
df_India_Times_articles = clean_dataframe(df_India_Times_articles)
df_NYT_cleaned = clean_dataframe(df_NYT_cleaned)

# Combine all dataframes by rows
df_all_newspapers = pd.concat([df_guardian_cleaned, df_ABC_headlines, df_Pakistan_headlines,
                               df_Irish_Times_articles, df_India_Times_articles, df_NYT_cleaned], ignore_index=True)

# Display the first few rows of the combined dataframe to inspect it
print(df_all_newspapers)
df_all_newspapers = df_all_newspapers.dropna()

print(df_all_newspapers)

# Save as Excel file because csv file made hella issues that took long to solve. I just need to remmeber to delete missing observations
excel_file_path = 'all_newspapers.xlsx'
df_all_newspapers.to_excel(excel_file_path, index=False)
```

```python
from transformers import RobertaTokenizer, RobertaModel
import torch

tokenizer = RobertaTokenizer.from_pretrained('roberta-base')
roberta_model = RobertaModel.from_pretrained('roberta-base')
roberta_model.eval()
roberta_model.to('cpu')  # Assuming you are using a CUDA-enabled device

def get_roberta_embeddings(text):
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=512)
    inputs = {key: value.to('cpu') for key, value in inputs.items()}
    with torch.no_grad():
        outputs = roberta_model(**inputs)
    # Use the pooled output for a single vector per input sequence
    return outputs.last_hidden_state[:, 0, :]  # Get the [CLS] token's embeddings
```

```python
class LSTMClassifier(torch.nn.Module):
    def __init__(self, embedding_dim, hidden_dim, output_dim):
        super(LSTMClassifier, self).__init__()
        self.lstm = torch.nn.LSTM(embedding_dim, hidden_dim, batch_first=True)
        self.fc = torch.nn.Linear(hidden_dim, output_dim)

    def forward(self, x):
        x = x.unsqueeze(0)  # Ensure there's a batch dimension
        _, (hn, _) = self.lstm(x)
        return self.fc(hn.squeeze(0))

# Load your trained model
def load_lstm_model(model_path, device='cpu'):
    model = LSTMClassifier(embedding_dim=768, hidden_dim=128, output_dim=2)  # Adjust dimensions as necessary
    model.load_state_dict(torch.load(model_path, map_location=torch.device('cpu')))
    model.to(device)
    model.eval()
    return model

lstm_model = load_lstm_model('best_model_roberta.pth')
```

```python
import torch
import numpy as np

def predict_headline(lstm_model, text):
    embeddings = get_roberta_embeddings(text)
    # Ensure embeddings are unsqueezed if not already batched
    if embeddings.dim() == 1:
        embeddings = embeddings.unsqueeze(0)
    output = lstm_model(embeddings)
    # Check output shape
    print("Output shape:", output.shape)
    scores = torch.softmax(output, dim=1).detach().cpu().numpy()
    return scores

# Define a function to add clickbait and non-clickbait scores as new columns
def add_scores_to_dataframe(df, lstm_model):
    clickbait_scores = []
    non_clickbait_scores = []
    for headline in df['Headline']:
        scores = predict_headline(lstm_model, headline)
        clickbait_scores.append(scores[0][0])
        non_clickbait_scores.append(scores[0][1])
    df['Clickbait Score'] = clickbait_scores
    df['Non-Clickbait Score'] = non_clickbait_scores

# Example Usage
add_scores_to_dataframe(df_all_newspapers, lstm_model)

# Display the updated DataFrame
print(df_all_newspapers)

df_all_newspapers_with_scores = df_all_newspapers

# Save as Excel file because csv file made hella issues that took long to solve. I just need to remmeber to delete missing observations
#excel_file_path = 'all_newspapers_with_scores.xlsx'
#df_all_newspapers.to_excel(excel_file_path, index=False)

```

```python
import pandas as pd
from scipy import stats

# Assuming df_all_newspapers_with_scores is your DataFrame containing clickbait scores
# Assuming the column 'Date' contains the dates and 'Clickbait Score' contains the clickbait scores
# Replace these column names with the actual ones in your DataFrame

# Convert the 'Date' column to datetime format
df_all_newspapers_with_scores['Date'] = pd.to_datetime(df_all_newspapers_with_scores['Date'])

# Drop rows with missing values in the 'Clickbait Score' column
df_all_newspapers_with_scores = df_all_newspapers_with_scores.dropna(subset=['Clickbait Score'])

# Filter the DataFrame to include only observations between 2000 and 2020
df_filtered = df_all_newspapers_with_scores[(df_all_newspapers_with_scores['Date'].dt.year >= 2000) & 
                                            (df_all_newspapers_with_scores['Date'].dt.year <= 2020)]

# Filter the DataFrame for observations before and after January 2014
before_2014 = df_filtered[df_filtered['Date'] < pd.to_datetime('2014-01-01')]['Clickbait Score']
after_2014 = df_filtered[df_filtered['Date'] >= pd.to_datetime('2014-01-01')]['Clickbait Score']

# Perform the Mann-Whitney U Test
u_statistic, p_value = stats.mannwhitneyu(before_2014, after_2014, alternative='two-sided')

# Print the results
print("Mann-Whitney U Test Results:")
print("U-statistic:", round(u_statistic, 3))
print("P-value:", round(p_value, 7))

# Interpret the results
alpha = 0.05  # Set your desired significance level
if p_value < alpha:
    print("There is a statistically significant difference in clickbait scores between the periods overall.")
else:
    print("There is no statistically significant difference in clickbait scores between the periods overall.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Specify the file path
excel_file_path = 'all_newspapers_with_scores.xlsx'

# Read the Excel file into a dataframe
df_all_newspapers_with_scores = pd.read_excel(excel_file_path)

# Drop rows where 'Date' column is NaN
df_all_newspapers_with_scores = df_all_newspapers_with_scores.dropna(subset=['Date'])

# Filter out rows where 'Date' column is not a valid date
df_all_newspapers_with_scores = df_all_newspapers_with_scores[df_all_newspapers_with_scores['Date'].astype(str).str.match(r'\d{4}-\d{2}-\d{2}')]

# Convert 'Date' column to datetime
df_all_newspapers_with_scores['Date'] = pd.to_datetime(df_all_newspapers_with_scores['Date'])

# Get the unique sources
sources = df_all_newspapers_with_scores['Source'].unique()

# Loop through each source and create a plot
for source in sources:
    # Filter the dataframe for the current source
    df_source = df_all_newspapers_with_scores[df_all_newspapers_with_scores['Source'] == source]
    
    # Group by month and calculate the average clickbait score
    avg_clickbait_score_per_month = df_source.groupby(df_source['Date'].dt.to_period('M'))['Non-Clickbait Score'].mean().reset_index()
    
    # Convert 'Date' period to datetime for plotting
    avg_clickbait_score_per_month['Date'] = avg_clickbait_score_per_month['Date'].dt.to_timestamp()
    
    # Separate the data into two periods: before and after January 2014
    avg_before_2014 = avg_clickbait_score_per_month[avg_clickbait_score_per_month['Date'] <= pd.Timestamp('2014-01-01')]['Non-Clickbait Score'].mean()
    avg_after_2014 = avg_clickbait_score_per_month[avg_clickbait_score_per_month['Date'] > pd.Timestamp('2014-01-01')]['Non-Clickbait Score'].mean()
    
    # Find the index corresponding to 1 month before January 2014
    idx_one_month_before_2014 = avg_clickbait_score_per_month[avg_clickbait_score_per_month['Date'] == pd.Timestamp('2013-12-01')].index[0]
    
    # Plot
    plt.figure(figsize=(10, 6))
    
    # Plot the average clickbait score per month using scatterplot only
    sns.scatterplot(x='Date', y='Non-Clickbait Score', data=avg_clickbait_score_per_month, color='b', s=100, marker='o', label='Average Clickbait Score')
    
    # Plot horizontal line for average clickbait score before January 2014
    plt.axhline(y=avg_before_2014, color='g', linestyle='-', linewidth=2, label='Average before Jan 2014', xmin=0, xmax=idx_one_month_before_2014/len(avg_clickbait_score_per_month))
    
    # Plot horizontal line for average clickbait score after January 2014
    plt.axhline(y=avg_after_2014, color='m', linestyle='-', linewidth=2, label='Average after Jan 2014', xmin=idx_one_month_before_2014/len(avg_clickbait_score_per_month), xmax=1)
    
    # Add vertical line at January 2014
    plt.axvline(pd.Timestamp('2014-01-01'), color='r', linestyle='--', linewidth=2, label='January 2014')
    
    # Set y-axis range from 0 to 0.5
    plt.ylim(0, 0.3)
    
    # Add x gridlines
    plt.grid(True, which='both', axis='x')
    
    # Add y gridlines
    plt.grid(True, which='both', axis='y')
    
    # Add gridlines per year
    plt.gca().xaxis.set_major_locator(plt.matplotlib.dates.YearLocator())
    plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter("%Y"))
    
    # Some style
    plt.xlabel('Month')
    plt.ylabel('Average Clickbait Score')
    plt.title(f"Average clickbait score per month for {source}")
    
    plt.legend()
    plt.tight_layout()
    
    # Rotate x-axis labels for better readability
    plt.xticks(rotation=45)
    
    # Show the plot
    plt.show()
```
